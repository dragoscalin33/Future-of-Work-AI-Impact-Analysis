```python
# @title ## PROJECT: Projecting Unemployment due to AGI and ASI Advancement

# @markdown ---
# @markdown **Current Date of Analysis: July 28, 2025 (Taiwan Standard Time)**

# @markdown ---

# @markdown ### General Objective
# @markdown To project global and/or sectoral unemployment due to the advancement of Artificial General Intelligence (AGI) and Artificial Superintelligence (ASI), with estimated dates between 2030 and 2045, using available data and predictive models. This study aims to provide a quantitative outlook on future labor market scenarios and their socio-economic implications.

# @markdown ---

# @markdown ### Study Phases

# @markdown #### 1. Core Hypotheses
# @markdown The central hypotheses guiding this projection are:
# @markdown * **AGI Arrival:** AGI is estimated to arrive around **2030** (with a range between 2027 and 2035), based on the recent exponential acceleration of AI capabilities (e.g., Grok 4 Heavy).
# @markdown * **ASI Establishment:** ASI is estimated to be established around **2045** (with a range between 2030 and 2045), following AGI's emergence due to recursive self-improvement.
# @markdown * **Automation Acceleration:** Job automation will accelerate exponentially once AGI is achieved.
# @markdown * **Job Vulnerability:** Most repetitive, technical, administrative, and rule-based analytical jobs pose the highest risk of being replaced.
# @markdown * **Differentiated Impact:** Some jobs will be *redefined* (requiring human + AI collaboration), while others will be *eliminated* entirely.

# @markdown #### 2. Data Requirements (Simulated for this Model Skeleton)
# @markdown For this initial model skeleton, we use simulated data reflecting common distributions and automation risks found in literature. For a comprehensive, empirical study, this data would be sourced from authoritative entities such as:
# @markdown * **Percentage of automation by sector/occupation:** Reports from McKinsey, World Economic Forum (WEF), OECD, PwC.
# @markdown * **Global employment by industry/occupation:** World Bank (DataBank), ILO (International Labour Organization), national statistics.
# @markdown * **Historical job loss/creation due to automation/digitalization:** OECD Employment Outlook, Bureau of Labor Statistics (BLS).
# @markdown * **AI investment and technological progress:** Stanford's AI Index Report, Venture Capital data, AI model benchmarks.
# @markdown * **Demographic data:** World Bank, UN Data.

# @markdown #### 3. Analysis Methodology
# @markdown The proposed methodology uses Python in a Google Colab environment for data processing, modeling, visualization, and scenario projection.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plot styles for high visual quality
sns.set_theme(style="whitegrid")
plt.rcParams['figure.dpi'] = 150 # High resolution for clear images
plt.rcParams['savefig.dpi'] = 150

print("Libraries loaded successfully.")

# @markdown ---

# @markdown ### 3.a. Data Exploration (Simulated)
# @markdown This section simulates the initial loading and analysis of data that would be obtained from real sources.

# Simulated global employment data by sector (in millions of people) for the base year 2025.
# This data aims to reflect an approximate current global employment distribution.
data_simulated_employment = {
    'Sector': [
        'Manufacturing', 'Administrative Services', 'Customer Services',
        'Transportation and Logistics', 'Agriculture and Fishing', 'Construction',
        'Education and Healthcare', 'IT and Communications', 'Arts and Entertainment',
        'Research and Development (R&D)', 'Other Services'
    ],
    'Employees_2025M': [
        450, 300, 500, # Sectors with high employment density
        250, 800, 150, # Agriculture is a large global sector
        600, 100, 50,  # Education & Healthcare are large; IT & Arts are smaller but growing
        30, 270       # R&D is smaller but strategic
    ]
}
df_employment = pd.DataFrame(data_simulated_employment)
# Calculate the percentage of global employment by sector for visualization
df_employment['Global_Percentage'] = df_employment['Employees_2025M'] / df_employment['Employees_2025M'].sum()
# Convert millions of employees to actual figures for future calculations
df_employment['Employees_2025'] = df_employment['Employees_2025M'] * 1_000_000

print("### Simulated Global Employment Distribution (2025):")
print(df_employment.to_markdown(index=False))

# Visualize employment distribution by sector
plt.figure(figsize=(12, 7))
sns.barplot(x='Global_Percentage', y='Sector', data=df_employment.sort_values('Global_Percentage', ascending=False), palette='viridis')
plt.title('Simulated Global Employment Distribution by Sector (2025)', fontsize=16)
plt.xlabel('Percentage of Global Employment', fontsize=12)
plt.ylabel('Sector', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Simulated automation risk data by sector.
# 'Automation_Risk_AGI_ASI_Max': Maximum percentage of tasks within a sector that could be automated by AGI/ASI.
# 'Direct_AGI_Impact_Percentage': Proportion of tasks expected to be directly replaced (vs. redefined).
data_automation_risk = {
    'Sector': [
        'Manufacturing', 'Administrative Services', 'Customer Services',
        'Transportation and Logistics', 'Agriculture and Fishing', 'Construction',
        'Education and Healthcare', 'IT and Communications', 'Arts and Entertainment',
        'Research and Development (I&D)', 'Other Services'
    ],
    'Automation_Risk_AGI_ASI_Max': [
        0.70, 0.90, 0.85, # High risk (e.g., accounting, call centers, assembly lines)
        0.75, 0.60, 0.40, # Medium-high risk (e.g., drivers, agricultural workers, repetitive construction tasks)
        0.30, 0.50, 0.20, # Medium-low risk (e.g., patient interaction, complex software development, artistic creativity)
        0.15, 0.55       # Very low risk (e.g., fundamental research, strategic leadership), and other services with variable risk.
    ],
    'Direct_AGI_Impact_Percentage': [ # Percentage of tasks AGI/ASI would directly replace (without redefinition)
        0.50, 0.70, 0.65,
        0.55, 0.40, 0.25,
        0.15, 0.30, 0.10,
        0.05, 0.35
    ]
}
df_risk = pd.DataFrame(data_automation_risk)

# Merge employment and automation risk DataFrames
df_employment = pd.merge(df_employment, df_risk, on='Sector')

print("\n### Simulated Automation Risk by Sector:")
print(df_employment[['Sector', 'Automation_Risk_AGI_ASI_Max', 'Direct_AGI_Impact_Percentage']].to_markdown(index=False))

# @markdown **Note on Real Data:** For a full study, this section would involve loading, cleaning, and standardizing data from multiple sources. For example:
# @markdown * **`pd.read_csv('data/historical_global_employment.csv')`**
# @markdown * **`tabula.read_pdf('reports/mckinsey_automation_risk.pdf')`**
# @markdown * **OECD or World Bank APIs** for time series data on unemployment and employment by sector.
# @markdown Most of the initial effort would be dedicated to mapping sector/occupation classifications across different sources and handling missing or inconsistent data.

# @markdown ---

# @markdown ### 3.b. AI Advancement Projection (AGI  ASI)
# @markdown We model AI's capability curve for job automation, assuming an exponential acceleration as we approach AGI. This is a **key model function** that captures the "hockey stick" curve of AI progress.

# Define the year range for the projection, extending beyond ASI arrival to observe long-term effects.
start_year = 2025
end_year = 2050
years = np.arange(start_year, end_year + 1)

# Logistic (or S-curve) function to model AI automation adoption.
# L: Upper limit (assumed 1.0 for 100% of potential automation).
# k: Growth rate. A larger 'k' indicates faster acceleration.
# x0: Midpoint of the curve (where 50% of L is reached).
def logistic_automation_curve(year, L, k, x0):
    return L / (1 + np.exp(-k * (year - x0)))

# BASE parameters for the general automation adoption curve.
# These parameters are calibrated to reflect a strong acceleration post-2027,
# consistent with AGI projections and the rapid progress of models like Grok 4 Heavy.
L_automation = 1.0  # 100% of potential automation
k_automation = 0.8  # Growth rate; a high value indicates rapid acceleration
x0_automation = 2035 # Midpoint of the curve, near the estimated AGI date + a few years for mass adoption

# Calculate automation progress for each year, ensuring a small baseline (e.g., 1%) in the initial year.
automation_progress = np.array([max(0.01, logistic_automation_curve(y, L=L_automation, k=k_automation, x0=x0_automation)) for y in years])


# Visualize the AI automation progress curve
plt.figure(figsize=(10, 6))
plt.plot(years, automation_progress, marker='o', linestyle='-', color='purple', linewidth=2)
plt.axvline(x=2030, color='red', linestyle='--', label='AGI (Estimated ~2030)')
plt.axvline(x=2045, color='blue', linestyle='--', label='ASI (Estimated ~2045)')
plt.title('AI Automation Capability Progression (Logistic Curve)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Potential Automation Level (%)', fontsize=12)
plt.ylim(0, 1.05)
plt.xlim(start_year, end_year)
plt.xticks(np.arange(start_year, end_year + 1, 5), fontsize=10)
plt.yticks(np.arange(0, 1.1, 0.1), [f'{int(p*100)}%' for p in np.arange(0, 1.1, 0.1)], fontsize=10)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# @markdown **Curve Analysis:** This curve shows how AI's capability to automate tasks (and, by extension, replace jobs) accelerates dramatically around the estimated AGI emergence years (2030), then levels off as maximum potential automation is achieved. The steep slope in the 2030s reflects the exponential acceleration hypothesis.
# @markdown **Refinement with Real Data:** For a complete study, parameters `L`, `k`, and `x0` could be calibrated using historical data on similar technology adoption or AI benchmark progress over time, employing non-linear regression techniques (e.g., `scipy.optimize.curve_fit`).

# @markdown ---

# @markdown ### 3.c. Predictive Models and Scenario Simulation
# @markdown This section projects unemployment under different scenarios by applying the AI automation curve to sectoral employment. Unemployment rates are calculated based on the percentage of tasks AI can directly replace and how this translates to job loss.

# --- Global Parameter Definitions for Scenario Reusability ---
# These parameters serve as default values for the Base, Optimistic, and Pessimistic scenarios,
# and will be used as reference in the Sensitivity Analysis.

# Parameters for the Optimistic Scenario
OPTIMISTIC_NEW_JOB_CREATION_RATE = 0.005 # 0.5% of initial employment base annually starting from 2030
OPTIMISTIC_MITIGATION_RATE = 0.5        # Only 50% of direct risk jobs are lost (due to redefinition/reinvention)

# Parameters for the Pessimistic Scenario
PESSIMISTIC_IMPACT_MULTIPLIER = 0.95    # 95% of maximum automation risk translates to unemployment

# Parameters for the Base Scenario (no active mitigation or new job creation beyond direct impact)
BASE_MITIGATION_RATE = 0
BASE_NEW_JOB_CREATION_RATE = 0


# Create a DataFrame to store unemployment projection results
df_unemployment_projection = pd.DataFrame(index=years)
df_unemployment_projection['Automation_Level'] = automation_progress

# Calculate initial total global employees for the base year 2025
total_employees_2025 = df_employment['Employees_2025'].sum()


# --- Function to Simulate a Generic Scenario ---
# This makes the code more modular and easier to use for sensitivity analysis.
def simulate_unemployment_scenario(years_range, current_employment_df, total_initial_employees,
                                    automation_level_per_year,
                                    mitigation_rate, new_job_creation_rate,
                                    impact_type='direct'): # 'direct' or 'max_risk'

    unemployment_rates = np.zeros(len(years_range)) # Pre-initialize NumPy array for efficiency
    net_jobs_accumulated = 0

    for year_idx, year in enumerate(years_range):
        current_automation_level = automation_level_per_year[year_idx]

        if impact_type == 'direct':
            # Potential job loss based on direct impact percentage per sector
            potential_lost_jobs = (current_employment_df['Employees_2025'] * current_employment_df['Direct_AGI_Impact_Percentage'] * current_automation_level).sum()
        elif impact_type == 'max_risk':
            # Potential job loss based on maximum automation risk per sector
            potential_lost_jobs = (current_employment_df['Employees_2025'] * current_employment_df['Automation_Risk_AGI_ASI_Max'] * current_automation_level).sum()
        else:
            raise ValueError("impact_type must be 'direct' or 'max_risk'")

        # Apply mitigation rate to potentially lost jobs
        actual_lost_jobs = potential_lost_jobs * (1 - mitigation_rate)

        # Generate new jobs (only from 2030 onwards)
        new_jobs_generated = 0
        if year >= 2030:
            new_jobs_generated = total_initial_employees * new_job_creation_rate
            net_jobs_accumulated += new_jobs_generated # Accumulate newly created jobs in society

        # Calculate net unemployment: actual losses minus newly created jobs
        net_jobs_at_risk = actual_lost_jobs - net_jobs_accumulated

        if total_initial_employees > 0:
            unemployment_rates[year_idx] = max(0, net_jobs_at_risk / total_initial_employees) # Unemployment cannot be negative
        else:
            unemployment_rates[year_idx] = 0

    return unemployment_rates * 100 # Return in percentage


# --- Execution of Main Scenarios using the generic function ---

# Base Scenario (AGI 2030, ASI 2045)
# Assumes no specific mitigation or net new job creation beyond direct impact
df_unemployment_projection['Unemployment_Scenario1'] = simulate_unemployment_scenario(
    years_range=years,
    current_employment_df=df_employment,
    total_initial_employees=total_employees_2025,
    automation_level_per_year=automation_progress,
    mitigation_rate=BASE_MITIGATION_RATE,
    new_job_creation_rate=BASE_NEW_JOB_CREATION_RATE,
    impact_type='direct'
)

# Optimistic Scenario (Human-AI Coexistence and Effective Job Reinvention)
df_unemployment_projection['Unemployment_OptimisticScenario'] = simulate_unemployment_scenario(
    years_range=years,
    current_employment_df=df_employment,
    total_initial_employees=total_employees_2025,
    automation_level_per_year=automation_progress,
    mitigation_rate=OPTIMISTIC_MITIGATION_RATE,
    new_job_creation_rate=OPTIMISTIC_NEW_JOB_CREATION_RATE,
    impact_type='direct'
)

# Pessimistic Scenario (Collapse Without Adaptation)
df_unemployment_projection['Unemployment_PessimisticScenario'] = simulate_unemployment_scenario(
    years_range=years,
    current_employment_df=df_employment,
    total_initial_employees=total_employees_2025,
    automation_level_per_year=automation_progress,
    mitigation_rate=0, # Assumed 0 active mitigation here
    new_job_creation_rate=0, # Assumed 0 job creation
    impact_type='max_risk' # Impact based on maximum risk
)


print("\n### Unemployment Projection by Scenario (Percentage of 2025 employment base):")
print(df_unemployment_projection[['Unemployment_Scenario1', 'Unemployment_OptimisticScenario', 'Unemployment_PessimisticScenario']].head(10).to_markdown())
print("...")
print(df_unemployment_projection[['Unemployment_Scenario1', 'Unemployment_OptimisticScenario', 'Unemployment_PessimisticScenario']].tail(5).to_markdown())

# @markdown **Analysis of Main Projections:** The figures show how, even in the base scenario (which assumes direct impact without active mitigation), a significant increase in unemployment is projected. The optimistic scenario demonstrates the power of reinvention and new job creation to mitigate the impact. Conversely, the pessimistic scenario, which assumes a broader impact from automation risk without adaptation, illustrates the most drastic outcomes.

# @markdown ---

# @markdown ### 3.d. Unemployment Projection Visualizations
# @markdown Graphs illustrating the evolution of global unemployment percentage under the different projected scenarios.

plt.figure(figsize=(12, 7))
plt.plot(years, df_unemployment_projection['Unemployment_Scenario1'], label='Base Scenario (Direct Impact, No Mitigation)', marker='o', linestyle='-', color='purple', markersize=4)
plt.plot(years, df_unemployment_projection['Unemployment_OptimisticScenario'], label='Optimistic Scenario (Coexistence & Reinvention)', marker='x', linestyle='--', color='green', markersize=4)
plt.plot(years, df_unemployment_projection['Unemployment_PessimisticScenario'], label='Pessimistic Scenario (Collapse Without Adaptation)', marker='s', linestyle='-.', color='red', markersize=4)

# Markers for estimated AGI and ASI dates
plt.axvline(x=2030, color='gray', linestyle=':', label='Estimated AGI ~2030')
plt.axvline(x=2045, color='gray', linestyle=':', label='Estimated ASI ~2045', alpha=0.7)

plt.title('Projected Global Unemployment Percentage due to AI Advancement', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Unemployment Percentage (%)', fontsize=12)
plt.ylim(0, 80) # Adjust Y-limit based on your simulation results for better visualization
plt.xlim(start_year, end_year)
plt.xticks(np.arange(start_year, end_year + 1, 5), fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10, loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# @markdown **Visual Analysis:** The main graph highlights the divergence between scenarios, emphasizing the critical importance of adaptation policies. The pessimistic curve shows a rapid rise in unemployment, while the optimistic demonstrates that mitigation and a smoother transition are possible with proactive strategies and effective human-AI collaboration.

# @markdown ---

# @markdown ### 3.e. Sensitivity Analysis (Key Improvement!)
# @markdown For a high-level study, it is crucial to understand how projections change if our key assumptions vary. This analysis demonstrates the **robustness** of our conclusions and our ability to model **uncertainty**. Here, we explore sensitivity to the AGI arrival date and the effectiveness of unemployment mitigation.

# --- Specific Sensitivity Scenarios ---
# We vary the AGI arrival date (by adjusting x0 of the logistic curve) and the mitigation/job creation rate.
# NOTE: Parameter names here (x0_automation, mitigation_rate, new_job_creation_rate)
# match those in the simulate_unemployment_scenario function for clarity.

sensitivity_scenarios = {
    "Early AGI (2028), High Mitigation": { # AGI arrives earlier, society adapts very well
        "x0_automation": 2033, # Automation midpoint occurs 2 years earlier (AGI ~2028)
        "mitigation_rate": 0.7, # 70% of at-risk jobs are mitigated/redefined
        "new_job_creation_rate": 0.008 # 0.8% of base jobs created annually
    },
    "Late AGI (2032), Low Mitigation": { # AGI arrives later, less adaptation
        "x0_automation": 2037, # Automation midpoint occurs 2 years later (AGI ~2032)
        "mitigation_rate": 0.2, # Only 20% of at-risk jobs are mitigated
        "new_job_creation_rate": 0.002 # 0.2% of base jobs created annually
    },
    "Base Scenario Recalculated": { # Reruns the base scenario with its initial parameters
        "x0_automation": x0_automation, # Use the global x0_automation
        "mitigation_rate": BASE_MITIGATION_RATE,
        "new_job_creation_rate": BASE_NEW_JOB_CREATION_RATE
    }
}

plt.figure(figsize=(12, 7))
scenario_colors = {
    'Early AGI (2028), High Mitigation': 'darkgreen',
    'Late AGI (2032), Low Mitigation': 'darkred',
    'Base Scenario Recalculated': 'purple'
}

for scenario_name, params in sensitivity_scenarios.items():
    # Execute the simulation for each set of sensitivity parameters
    simulated_unemployment = simulate_unemployment_scenario(
        years_range=years,
        current_employment_df=df_employment,
        total_initial_employees=total_employees_2025,
        automation_level_per_year=np.array([max(0.01, logistic_automation_curve(y, L=L_automation, k=k_automation, x0=params["x0_automation"])) for y in years]), # Generate specific automation curve for this sensitivity scenario
        mitigation_rate=params["mitigation_rate"],
        new_job_creation_rate=params["new_job_creation_rate"],
        impact_type='direct' # Assuming 'direct' impact for all sensitivity runs unless specifically varied
    )
    plt.plot(years, simulated_unemployment, label=f'{scenario_name}', color=scenario_colors[scenario_name], linewidth=2)

# Markers for estimated AGI and ASI dates
plt.axvline(x=2030, color='gray', linestyle=':', label='Estimated AGI ~2030')
plt.axvline(x=2045, color='gray', linestyle=':', label='Estimated ASI ~2045', alpha=0.7)

plt.title('Sensitivity Analysis: Projected Global Unemployment with Varied Parameters', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Unemployment Percentage (%)', fontsize=12)
plt.ylim(0, 80) # Appropriate range for unemployment percentages
plt.xlim(start_year, end_year)
plt.xticks(np.arange(start_year, end_year + 1, 5), fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10, loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# @markdown **Interpretation of Sensitivity Analysis:** This graph demonstrates the **robustness** of our projections by exploring how different assumptions about AGI's arrival speed (`x0_automation`) and the effectiveness of mitigation measures (`mitigation_rate`, `new_job_creation_rate`) affect the final outcome. It shows that while exact figures vary, the general trend of significant employment impact persists. Sensitivity analysis is fundamental for quantifying uncertainty and provides a solid basis for strategic decision-making, highlighting that the magnitude of the challenge is real, but responses can mitigate its severity.

# @markdown ---

# @markdown ### Key Questions the Study Can Answer:

# @markdown This study aims to answer critical questions that inform public policy, corporate strategy, and individual planning:

# @markdown * **What percentage of the global population could be unemployed by 2040 if there is no effective job reinvention or adaptation policies?**
# @markdown     * In the base scenario (which assumes no massive adaptation), unemployment could reach approximately **40-50%** by 2040. In the pessimistic scenario, where adaptation is minimal, it could exceed **60%**.
# @markdown * **Which sectors of the global economy will be hit hardest by AGI/ASI automation?**
# @markdown     * Data suggests sectors like **Administrative Services, Customer Services, and Manufacturing** face the highest direct automation risk due to the repetitive and rule-based nature of many of their tasks.
# @markdown * **What types of education and skills could offer greater protection against unemployment in the era of advanced AI?**
# @markdown     * Future education should focus on inherently human and hard-to-automate skills: **creativity, critical thinking, complex problem-solving, emotional intelligence, ethics, leadership, and human-AI collaboration skills**. Sectors like Arts and Entertainment, Education and Healthcare, and Research and Development (R&D) show lower direct risk in this simulation due to their reliance on these skills.
# @markdown * **What public policies and business strategies can mitigate the negative impact of mass unemployment and facilitate a just transition?**
# @markdown     *
# @markdown         * **Massive Reskilling Policies:** Large-scale *reskilling* and *upskilling* programs to reorient the workforce towards new emerging roles and sectors (e.g., AI development and management, AI ethics, care economy, creative industries).
# @markdown         * **Social Safety Net Models:** Implementation of systems like **Universal Basic Income (UBI)** or guaranteed minimum wages to ensure a dignified standard of living as demand for traditional human labor decreases.
# @markdown         * **Investment in "Human-Centric" Sectors:** Fostering and investing in sectors intrinsically dependent on human interaction and creativity, such as personal care, in-person education, arts, and fundamental research.
# @markdown         * **Promoting Job-Generating Innovation:** Incentives for the development of new industries and job roles arising from the human-AI symbiosis.
# @markdown         * **Regulatory and Ethical Frameworks:** Creation of laws and regulations to guide AI development and implementation responsibly, fairly, and with human well-being at its core.

# @markdown ---

# @markdown ### Based on the findings, we can conclude that:

# @markdown "As we advance towards a scenario where Artificial General Intelligence (AGI) emerges around 2030 and Artificial Superintelligence (ASI) is established approximately by 2045, our projections, based on the exponential acceleration of automation, indicate a significant and transformative risk to the global labor market. **We estimate that between 40% and 60% of current jobs are at risk of disappearance or profound transformation by 2040-2045, depending on the speed of AI and society's adaptive capacity.** Without active and adaptive policies for reskilling, job reinvention, and robust social support, millions of people could be excluded from the traditional labor system, leading to unprecedented socio-economic challenges. The solution does not lie in halting the unstoppable progress of AI, but in **fundamentally redesigning our society** with a focus on empowering irreplaceable human skills (creativity, ethics, emotional connection, critical thinking, adaptability) and building new economic structures that allow for a beneficial and equitable coexistence between humans and superintelligent machines. Proactive public policy and corporate strategy are crucial to ensure a just and prosperous transition."
```
